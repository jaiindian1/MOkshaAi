```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
cancer = pd.read_csv('/content/breast_cancer.csv')
cancer.head()
cancer.info()
cancer.describe()
```

```python
cancer = cancer.drop(['Unnamed: 32', 'id'], axis=1)
```

```python
cancer.head()
```

```python
cancer.info()
```

```python
cancer['diagnosis'] =cancer['diagnosis'].replace({'M':1,'B':0})
```

```python
cancer.info()
```

```python
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
Y = cancer['diagnosis']
X = cancer.drop(['diagnosis'],axis =1)
x_train,x_test,y_train,y_test =train_test_split(X,Y,test_size=0.2 ,random_state=42)
model = SVC()
```

```python
model =model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print(y_test,y_pred)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001]
}

# Scale the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

grid = GridSearchCV(
    estimator=SVC(),
    param_grid=param_grid,
    verbose=3
)

grid.fit(x_train_scaled, y_train)

# Print the best parameters found
print("\n--- Grid Search Results ---")
print("Best Parameters found on training set:", grid.best_params_)
print("Best Score (Accuracy) found on training set:", grid.best_score_)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Use the best model found by the grid search for predictions
final_predictions = grid.best_estimator_.predict(x_test_scaled)

print("\n--- Final Tuned Model Evaluation on Test Set ---")

# Print the Confusion Matrix (Corrected Code)
print("Confusion Matrix:")
print(confusion_matrix(y_test, final_predictions))

# Print the Classification Report
print("\nClassification Report:")
print(classification_report(y_test, final_predictions))
```

```python
import joblib

# 1. Define the filename for the model
model_filename = 'tuned_svm_model.joblib'

# 2. Save the best model estimator object to the file
joblib.dump(grid.best_estimator_, model_filename)

print(f"Model saved locally as: {model_filename}")
```
